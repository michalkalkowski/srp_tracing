```python
%matplotlib notebook
import pogopy.misc as misc
```

```python
c_weld = 1e9*np.array([[234, 118, 148, 0, 0, 0],
                       [118, 240, 146, 0, 0, 0],
                       [148, 146, 210, 0, 0, 0],
                       [0, 0, 0, 99, 0, 0],
                       [0, 0, 0, 0, 110, 0],
                       [0, 0, 0, 0, 0, 95]])

rho_weld = 8.0e3
phase_angles = np.linspace(0, 2*np.pi, 200)
```

```python
cp, m, cg, _ = misc.calculate_slowness(c_weld, rho_weld, phase_angles)
```

```python
plt.polar(phase_angles, cp[:, 0], label='phase velocity')
plt.polar(phase_angles, np.linalg.norm(cg, axis=1)[:, 0], label='group velocity')
plt.legend()
```

The energy travels along the direction dictated by the group velocity vector. We may then plot the group velocity as a function of the ray angle.

```python
plt.figure()
plt.polar(phase_angles, np.linalg.norm(m, axis=1)[:, 0])
```

```python
ray_angle = np.arctan2(cg[:, 2, 0].real, cg[:, 1, 0].real)
ray_angle[ray_angle < 0] += 2*np.pi
plt.figure()
plt.polar(phase_angles, cp[:, 0], label='phase velocity')
plt.polar(phase_angles, np.linalg.norm(cg, axis=1)[:, 0], label='against phase angle')
plt.polar(ray_angle, np.linalg.norm(cg, axis=1)[:, 0], label='against ray angle')
#plt.polar(ray_angle - skew_angle, np.linalg.norm(cg, axis=1)[:, 0], label='against ray angle')
plt.legend()
```

Group velocity vector is perpendicular to the tangent of the phase slowness curve. For isotropic media the ray angle is the same as the phase angle, but for anisotropic media there is a skew angle between the two, dependent on the level and form of anisotropy. Group velocity vector and the phase velocity are related through the skew angle $\phi$:
$$
c_g \cos{\phi} = c_p
$$
from where the skew angle can be calculated as
$$
\phi = \arccos {\frac{c_p}{c_g}}
$$

If we denote phase angle as $\theta$, then ray angle can be defined as $\theta + \phi$. However, one needs to acount for the fact that $\phi$ is sometimes positive and sometimes negative, which is not captured in the expression above. This is not discussed further here.

```python
skew_angle = np.arccos(cp[:, 0]/np.linalg.norm(cg, axis=1)[:, 0].real)
```

```python
plt.figure()
plt.plot(phase_angles, phase_angles)
plt.plot(phase_angles, ray_angle)
```
